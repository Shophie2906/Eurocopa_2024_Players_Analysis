```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Prep_Airs_Flow_Card.csv')
```

```python
df.head()
```

```python
# Delimiter is '//'
df[['Date', 'Flight Number', 'From_To', 'Class', 'Price']] = df['Flight Details'].str.split('//', expand=True)
df[['From', 'To']] = df['From_To'].str.split('-', expand=True)
df = df.drop(columns=['From_To'])
```

```python
# Convert 'DATE' into date format
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# Convert 'Price' to a decimal value
df['Price']= pd.to_numeric(df['Price'], errors='coerce')
df['Price'] = df['Price'].round(2)
```

```python
# Change the Flow Card field to Yes / No values instead of 1 / 0
df['Flow Card?']=pd.to_numeric(df['Flow Card?'], errors='coerce')
df['Flow Card?']= df['Flow Card?'].replace(1, 'Yes')
df['Flow Card?']= df['Flow Card?'].replace(0, 'No')
```

```python
# Both outputs have the same data structure
new_order = ['Date', 'Flight Number', 'From', 'To', 'Class', 'Price', 'Flow Card?', 'Bags Checked', 'Meal Type']
df = df[new_order]
```

```python
# Create two tables, one for Flow Card holders and one for non-Flow Card holders
df_yes = df[df['Flow Card?'] == 'Yes']
df_no = df[df['Flow Card?'] == 'No']
```
